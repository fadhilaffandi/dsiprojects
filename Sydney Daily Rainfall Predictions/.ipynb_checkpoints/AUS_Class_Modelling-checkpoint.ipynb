### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, roc_auc_score

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
aus_rain = pd.read_csv('./datasets/encoded_aus_weather.csv')
```

# Train-test-split

```python
features = aus_rain.columns.tolist()
features.remove("RainTomorrow")
features.remove("Date")

X = aus_rain[features]
y = aus_rain['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

# List of Classifiers

- Logistic Regression
- ~~Naive Bayes~~ (usually used for discrete features) and we have negative continuous values NB is unable to cope with
- Random Forest
- Extra Trees
- GradientBoost
- AdaBoost
- SVC
- XGBoost
- Neural Networks

# Logreg

```python
lr = LogisticRegression()
lr.fit(X_train_sc, y_train)
train_acc = lr.score(X_train_sc,y_train)
print(f"Train accuracy is {train_acc}.")

est_test_acc = cross_val_score(lr, X_train_sc, y_train, cv=5).mean()
print(f"Estimated Test accuracy is {est_test_acc}.")

test_acc = lr.score(X_test_sc,y_test)
print(f"Test accuracy is {test_acc}.")
```

```python
y_pred = lr.predict(X_test_sc)
y_pred_proba = lr.predict_proba(X_test_sc)
roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
roc_auc
```

```python
coeff = lr.coef_
feature_importance = pd.Series(coeff[0], index=features)
feature_importance = feature_importance.sort_values()
feature_importance.plot.barh(figsize=(10,26))
```

```python
feature_importance.abs().sort_values(ascending=False).head(15)
```

Our top 5 features are Pressure3pm, Humidity3pm, Pressure9am, WindGustSpeed and MinTemp.

# Random Forest

```python
rf = RandomForestClassifier()
rf.fit(X_train_sc, y_train)
train_acc = rf.score(X_train_sc,y_train)
print(f"Train accuracy is {train_acc}.")

est_test_acc = cross_val_score(rf, X_train_sc, y_train, cv=5).mean()
print(f"Estimated Test accuracy is {est_test_acc}.")

test_acc = rf.score(X_test_sc,y_test)
print(f"Test accuracy is {test_acc}.")
```

```python
y_pred = rf.predict(X_test_sc)
y_pred_proba = rf.predict_proba(X_test_sc)
roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
roc_auc
```

```python
feature_names_sorted = np.take(X_train.columns, rf.feature_importances_.argsort()[::-1])
feature_imp_sorted = np.sort(rf.feature_importances_)[::-1]

# Plotting barplot of features and the logreg coefs.

plt.figure(figsize=(10,26))

sns.barplot(orient='h', x=feature_imp_sorted, y=feature_names_sorted, color='royalblue')

# plt.tick_params(axis='x', labelrotation=25)

# Setting the size of axis ticks.
plt.tick_params(axis='x', labelsize=12)
plt.tick_params(axis='y', labelsize=10)

plt.xlabel('Feature Importance', fontsize=14)
plt.ylabel('Features', fontsize=14)

plt.title("Barplot of Feature Importances", fontsize=18)

# Exporting plot image in PNG format.
#plt.savefig('../plot_images/barplot_rfc_feature_importance.png', bbox_inches='tight');
```

# Extra Trees

```python
et = ExtraTreesClassifier()
et.fit(X_train_sc, y_train)
train_acc = et.score(X_train_sc,y_train)
print(f"Train accuracy is {train_acc}.")

est_test_acc = cross_val_score(et, X_train_sc, y_train, cv=5).mean()
print(f"Estimated Test accuracy is {est_test_acc}.")

test_acc = et.score(X_test_sc,y_test)
print(f"Test accuracy is {test_acc}.")
```

```python
y_pred = et.predict(X_test_sc)
y_pred_proba = et.predict_proba(X_test_sc)
roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
roc_auc
```
