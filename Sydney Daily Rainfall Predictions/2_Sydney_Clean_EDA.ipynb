## Distribution of Features

```python
#function to create histograms for the distribution of features
def subplot_histograms(plotname, dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(15,60)) # specified figsize
    fig.suptitle(plotname, fontsize=16)# plot title added
    fig.subplots_adjust(top=0.95)
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column])
        ax[i].set_title(column)  #title for each plot
        ax[i].set_xlabel(column, size = 'large') #xlabel
        ax[i].set_ylabel('Count', size= 'large')  #ylabel
```

```python
#removing date from our list of features
l1 = sydney.columns.tolist()
l2 = ['Date']
syd_cols = [x for x in l1 if x not in l2]
```

```python
subplot_histograms('Distribution of Features',sydney,
                   syd_cols)
#plt.savefig('./images/syd_dist_of_features.png', bbox_inches='tight')
```

Our distributions are similar to that of Australia's with the exception of Wind Direction.  


It appears that for Sydney, there's significantly alot less Northern winds at both measurement times and a Huge amount of Southern Winds at 9am.The morning wind directionss could be related to the high tides present. This could be attributed to its location being the Eastern side of Australia, by the Pacific Ocean.  


As for Cloud9am/3pm, there seems to be less of the 2 to 5 okta regions of coverage. This means that for Sydney, there's either next to no clouds, or alot of it at once. Which in turn corresponds to the distribution of Sunshine that Sydney has. The Sunshine peaks out at 10hours and the 2nd highest is 0 hours.

## Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(15,8)) #creating figure area
ax.set_title('Correlation Heatmap of Sydney Rain') #setting title
mask = np.zeros_like(sydney.corr())
mask[np.triu_indices_from(mask)] = True
ax = sns.heatmap(sydney.corr(), annot=True,mask=mask )  #plotting heatmap
bottom, top = ax.get_ylim() #add this
ax.set_ylim(bottom + 0.5, top - 0.5)
```

Again we see collinear variables but there are no singular feature that jumps out when it comes to Rainfall. The highest correlation value with respect to Rainfall that can be found is Humidity at around 0.3.

```python
#splitting the date to day, month, year for better visualisations
sydney['month'] = sydney['Date'].dt.month
sydney['year']= sydney['Date'].dt.year
sydney['day']= sydney['Date'].dt.day
```

```python
#acquiring all our nominal features
obj_typed = sydney.select_dtypes(include='object').columns.values.tolist()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=1, figsize=(8,8))
axes = axes.ravel()
for i, object_t in enumerate(obj_typed):
    box = sns.scatterplot(x=object_t, y='Rainfall', data=sydney , ax= axes[i])
    x_ticks = sydney[object_t].unique()
    box.set_xticklabels(labels = x_ticks, rotation = 90 )
fig.tight_layout(pad=2.0)
```

## Pairplots

```python
sns.pairplot(data = sydney, x_vars='Cloud9am', y_vars='Rainfall', hue='month', height=5);
```

This is quite intuitive as one would expect there to be more Rainfall when the Clouds are bigger, and bigger clouds cover more of the sky.  
Therefore the more Cloud coverage, the more Rainfall.  
It also appears that there's lesser Rainfall for November and December.

```python
sns.pairplot(data = sydney, x_vars='Evaporation', y_vars='Rainfall', height=5);
```

Surprisingly, there seems to be more Rainfall with less Evaporation. It could be that Evaporation is an indicator of something else.

```python
sns.pairplot(data = sydney, x_vars='Evaporation', y_vars='Humidity9am', height=5);
```

There is a lot less Evaporation when there is more Humidity. Given that Humidity is a measure of moisture saturation in the air, it makes sense that with more saturation, there will be less evaporation adding moisture on to that saturated system.

```python
plt.figure(figsize = (10,10))
sns.pairplot(data = sydney, x_vars='Sunshine', y_vars='Rainfall', hue='Cloud9am', palette = 'husl', height=5);
#plt.savefig('./images/syd_rain_sun_cloud.png', bbox_inches='tight')
```

Again we see the inverse correlation between Sunshine and Cloud coverage. Also the reiteration of the positively correlation between Cloud Coverage and Rainfall, or in this case, the inverse correlation between Sunshine and Rainfall.

## Bargraphs

```python
mth_rain = sydney.pivot_table(index="Cloud9am", values="Sunshine")
mth_rain_sorted = mth_rain.sort_values(by=["Sunshine"])
mth_rain_sorted.plot.barh(figsize=(8,6), color= 'skyblue')
plt.title('Cloud vs Sunshine')
plt.ylabel('Cloud Oktas')
plt.xlabel('Hours of Sunshine');
```

Well, here its quite obvious that the more Cloud Coverage, the less Sunshine there is.

```python
mth_rain = sydney.pivot_table(index="month", values="RainToday")
mth_rain_sorted = mth_rain.sort_values(by=["RainToday"])
mth_rain_sorted.plot.barh(figsize=(8,6), color= 'mediumturquoise')
plt.title('Monthly Probability')
plt.ylabel('Month')
plt.xlabel('Probability of Rain Today');
```

And, with RainToday being more than 1mm of Rainfall. Here we see June being the highest and April being second. It is surprising that May or July did not come second place instead.  
This shows that Rainfall does'nt exactly follow a simple periodic or cylical trend.
